```python
import numpy as np
import pandas as pd
```

```python
df_train = pd.read_csv('./training_data.csv')
df_validate = pd.read_csv('./validation_data.csv')
```

```python
Y = df_train.iloc[:,-1]
```

```python
X = df_train.iloc[:,:40]
```

```python
Y = np.asarray(Y)
X = np.asarray(X)
```

```python
df_train.head()
```

```python
df_validate.head()
```

```python
print(df_train.shape)
print(df_validate.shape)
```

```python
def sigmoid(w,x):
    return 1/(1+np.exp(-(w*x)))

def error(w):
    err = 0
    for x,y in zip(X,Y):
        fx = sigmoid(w,x)
        err = err+0.5*(fx-y)**2
    return err
```

```python
w = np.load('weights_after_epoch_0.npy')
err = error(w)
```

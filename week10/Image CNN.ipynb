```python
import numpy as np
import pandas as pd
import cv2
```

```python
w = pd.read_csv('Q8_image.csv')
```

```python
w.head
```

```python
w = np.asarray(w)
```

```python
cv2.imwrite("./unfilter.png",w)
```

```python
kernel = np.array([[0.053,0.110,0.054],
                   [0.111,0.225,0.111],
                   [0.054,0.110,0.053]])
```

```python
dst = cv2.filter2D(w,-1,kernel)
```

```python
cv2.imwrite("./filter.png",dst)
```
